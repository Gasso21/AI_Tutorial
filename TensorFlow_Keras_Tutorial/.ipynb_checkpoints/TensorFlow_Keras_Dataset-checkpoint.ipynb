## Preprocessing

```python
# Make the data scale distribution between 0 and 1
train_images = train_images.astype(np.float32)/255
test_images = test_images.astype(np.float32)/255

# One-Hot Encoding
train_labels = keras.utils.to_categorical(train_labels, 10)
test_labels = keras.utils.to_categorical(test_labels, 10)
```

## Make Dataset

```python
train_dataset = tf.data.Dataset.from_tensor_slices((train_images, train_labels)).shuffle(buffer_size=100000).batch(64)
test_dataset = tf.data.Dataset.from_tensor_slices((test_images, test_labels)).batch(64)
```

# Model
## Keras Sequential API1

```python
model1 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dropout(.2),
    keras.layers.Dense(10, activation='softmax'),
])
```

```python
model1.summary()
```

## Keras Sequential API2

```python
def create_seq_model():
    model2 = keras.Sequential()
    model2.add(keras.layers.Flatten(input_shape=(28,28))) # Multi-Layer Perceptron  1 
    model2.add(keras.layers.Dense(128, activation='relu')) # 128 Perceptron  layer
    model2.add(keras.layers.Dropout(.2)) # Overfitting 
    model2.add(keras.layers.Dense(10, activation='softmax')) # 10 Perceptron  layer
    return model2
```

```python
seq_model = create_seq_model()
```

```python
seq_model.summary()
```

\rightarrow \text{None: Batch Size}

## Keras Functional API

**Sequential api** vs **Functional api**

```python
def creat_func_model():
    inputs = keras.Input(shape=(28,28))
    flatten = keras.layers.Flatten()(inputs)
    dense = keras.layers.Dense(128, activation='relu')(flatten)
    drop = keras.layers.Dropout(.2)(dense)
    outputs = keras.layers.Dense(10, activation='softmax')(drop)
    model3 = keras.Model(inputs=inputs, outputs=outputs)
    return model3
```

```python
func_model = creat_func_model()
```

```python
func_model.summary()
```

## Model Class Subclassing
**Pytorch **

```python
class SubClassModel(keras.Model):
    def __init__(self):
        super(SubClassModel, self).__init__()
        self.flatten = keras.layers.Flatten(input_shape=(28,28))
        self.dense1 = keras.layers.Dense(128, activation='relu')
        self.drop = keras.layers.Dropout(.2)
        self.dense2 = keras.layers.Dense(10, activation='softmax')
    def call(self, x, training=False):
        x = self.flatten(x)
        x = self.dense1(x)
        x = self.drop(x)
        return self.dense2(x)
```

```python
subclass_model = SubClassModel()
```

```python
inputs = tf.zeros((1, 28, 28))
subclass_model(inputs)
subclass_model.summary()
```

```python
inputs = tf.random.normal((1,28,28))
outputs = subclass_model(inputs)
pred = tf.argmax(outputs, -1)
print(f"Predicted class: {pred}")
```

## Gradient Tape

```python
#loss function
loss_object = keras.losses.CategoricalCrossentropy()

#optimizer
learning_rate = 0.001
optimizer = keras.optimizers.Adam(learning_rate=learning_rate)

#loss, accuracy compute
train_loss = keras.metrics.Mean(name='train_loss')
train_accuracy = keras.metrics.CategoricalAccuracy(name='train_accuracy')

test_loss = keras.metrics.Mean(name='test_loss')
test_accuracy = keras.metrics.CategoricalAccuracy(name='test_accuracy')
```

```python
# decorator : graph mode  freeze; speed up,       
# debugging         
# keras api  graph mode 
@tf.function
def train_step(model, images, labels):
    with tf.GradientTape() as tape:
        # training=True is only needed if there are layers with different
        # behavior during training versus inference (e.g. Dropout).
        predictions = model(images, training=True) # 'call' method, training
        loss = loss_object(labels, predictions) # loss compute
    gradients = tape.gradient(loss, model.trainable_variables) # loss variables(weight, bias) 
    optimizer.apply_gradients(zip(gradients, model.trainable_variables)) #   variables  
    
    train_loss(loss)
    train_accuracy(labels, predictions)
```

```python
@tf.function
def test_step(model, images, labels):
    # training=False is only needed if there are layers with different
    # behavior during training versus inference (e.g. Dropout)
    predictions = model(images, training=False)
    t_loss = loss_object(labels, predictions)
    
    test_loss(t_loss)
    test_accuracy(labels, predictions)
```

```python
EPOCHS = 10

for epoch in range(EPOCHS):
    # Reset the metric at the start of the next epoch
    train_loss.reset_states()
    train_accuracy.reset_states()
    test_loss.reset_states()
    test_accuracy.reset_states()
    
    for images, labels in train_dataset:
        train_step(func_model, images, labels)
    
    for test_images, test_labels in test_dataset:
        test_step(func_model, test_images, test_labels)
        
    print(
        f'Epoch {epoch+1}, '
        f'Loss: {train_loss.result()}, '
        f'Accuracy: {train_accuracy.result()*100}. '
        f'Test Loss: {test_loss.result()}, '
        f'Test Accuracy: {test_accuracy.result()*100}'
    )
```
