## Partial Derivative

$$
    f(w_0,w_1) = w^2_0+2w_0w_1+3
$$

$$
    \nabla{_w}f =
        \left[ {\begin{array}{c}
            \cfrac{\partial{f}}{\partial{w_0}} \\
            \cfrac{\partial{f}}{\partial{w_1}}
        \end{array}} \right]
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(w0, w1): # 함수 f
    return w0**2 + 2*w0*w1 + 3

def df_dw0(w0, w1): # f의 w0에 대한 편미분
    return 2*w0 + 2*w1

def df_dw1(w0, w1): # f의 w1에 대한 편미분
    return 2*w0

w_range = 2 # 범위
dw = 0.25 # 간격
w0 = np.arange(-w_range, w_range +dw, dw)
#w0 = np.linspace(-w_range, w_range, 17)
w1 = np.arange(-w_range, w_range +dw, dw)
wn = w0.shape[0] # w0.shape = (17,)
xx0, xx1 = np.meshgrid(w0, w1) # 격자, 17 x 17
ff = np.zeros((len(w0), len(w1)))
dff_dw0 = np.zeros((len(w0), len(w1)))
dff_dw1 = np.zeros((len(w0), len(w1)))

for i in range(wn):
    ff[i,:] = f(w0[:], w1[i]) # 2차원 배열
    dff_dw0[i, :] = df_dw0(w0[:], w1[i])
    dff_dw1[i, :] = df_dw1(w0[:], w1[i])

# 등고선
plt.figure(figsize=(9, 4))
plt.subplots_adjust(wspace=0.3) # subplots 간격
plt.subplot(1, 2, 1)
cont = plt.contour(xx0, xx1, ff, 10, colors='black') # 등고선
cont.clabel(fmt='%2.0f', fontsize=8) # 등고선 눈금
plt.xticks(range(-w_range, w_range+1, 1)) # x축 눈금
plt.yticks(range(-w_range, w_range+1, 1)) # y축 눈금
plt.xlim(-w_range - .5, w_range + .5 ) # x축 범위
plt.ylim(-w_range - .5, w_range + .5 ) # y축 범위
plt.xlabel('$w_0$', fontsize=14)
plt.ylabel('$w_1$', fontsize=14)

# 경사도
plt.subplot(1, 2, 2)
plt.quiver(xx0, xx1, dff_dw0, dff_dw1) # f의 경사 벡터 표시
plt.xticks(range(-w_range, w_range+1, 1)) # x축 눈금
plt.yticks(range(-w_range, w_range+1, 1)) # y축 눈금
plt.xlim(-w_range - .5, w_range + .5 ) # x축 범위
plt.ylim(-w_range - .5, w_range + .5 ) # y축 범위
plt.xlabel('$w_0$', fontsize=14)
plt.ylabel('$w_1$', fontsize=14)

plt.show()
```

### 로그 함수의 최소값
$$
    \underset{x}{\operatorname{argmin}}f(x)
        = \underset{x}{\operatorname{argmin}}logf(x)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-4, 4, 100)
y = (x-1)**2 + 2
logy = np.log(y)

plt.figure(figsize = (4,4))
plt.plot(x, y, 'k', linewidth = 3)
plt.plot(x, logy, 'cornflowerblue', linestyle = '--', linewidth = 3)
plt.yticks(range(-4, 9, 1))
plt.xticks(range(-4, 5, 1))
plt.ylim(-4, 8)
plt.xlim(-4, 4)
plt.grid(1)
plt.show()

print('y의 최소값을 나타내는 x는', x[np.argmin(y)])
print('logy의 최소값을 나타내는 x는', x[np.argmin(logy)])
```

### Softmax function

$$
    u = exp(x_0) + exp(x_1) + exp(x_2)
$$
$$
    y_0 = \cfrac{exp(x_0)}{u},\quad y_1 = \cfrac{exp(x_1)}{u},\quad y_2 = \cfrac{exp(x_2)}{u}
$$

```python
import numpy as np
def softmax(x0, x1, x2):
    u = np.exp(x0) + np.exp(x1) + np.exp(x2)
    return np.exp(x0)/u, np.exp(x1)/u, np.exp(x2)/u 

y = softmax(2, 1, -1)
print(np.round(y,2))
print(np.sum(y))
```

- **softmax**
    - n개의 수의 대소 관계를 유지하면서 각각의 확률을 나타내는 경우

```python
# x2 = 1로 고정하고 x0과 x1을 입력했을 때, y0, y1을 플롯
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def softmax(x0, x1, x2):
    u = np.exp(x0) + np.exp(x1) + np.exp(x2)
    return np.exp(x0)/u, np.exp(x1)/u, np.exp(x2)/u

xn = 20
x0 = np.linspace(-4, 4, xn)
x1 = np.linspace(-4, 4, xn)
x2 = 1
y = np.zeros((xn, xn, 3))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0, :] = softmax(x0[i0], x1[i1], x2)
    
xx0, xx1 = np.meshgrid(x0, x1)
plt.figure(figsize = (10, 3))
for i in range(3):
    ax = plt.subplot(1, 3, i+1, projection = '3d')
    ax.plot_surface(xx0, xx1, y[:, :, i], rstride=1, cstride=1,alpha=0.3, color='blue', edgecolor='black')
    ax.set_xlabel('$x_0$', fontsize=14)
    ax.set_ylabel('$x_1$', fontsize=14)
    ax.view_init(40, -75)
    
plt.show()
```

### Softmax function & Sigmoid function
- 2변수의 경우 소프트맥스 함수
\begin{aligned}
    &y  = 
        \cfrac{e{^{x_0}}}{e{^{x_0}}+e{^{x_1}}} \\
    \rightarrow &y =
        \cfrac{e{^{x_0}}e{^{-{x_0}}}}{e{^{x_0}}e{^{-{x_0}}}+e{^{x_1}}e{^{-{x_0}}}}
        =
        \cfrac{e^{{x_0}-{x_0}}}{e^{{x_0}-{x_0}}+e^{{x_1}-{x_0}}}
        =
        \cfrac{1}{1+e^{-(x_0-x_1)}} \\
    &if \;x = x_0-x_1, \quad
    y =
        \cfrac{1}{1+e^{-x}}
\end{aligned}

### Gauss function

\begin{equation}
    y = a \; \exp \; \left( - \cfrac{(x- \mu)^2}{\sigma ^2} \right )\\
\begin{aligned}
    &\mu \; : \; 중심(평균) \\
    &\sigma \; : \; 확산(표준 \; 편차) \\
    &a \; : \; 높이 \\
    x로 \; 적분하여 \; 1로 \; 만들려면 \\
    a = \cfrac{1}{(2\pi\sigma{^2})^{\frac{1}{2}}}
\end{aligned}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def gauss(mu, sigma, a):
    return a*np.exp(-(x-mu)**2/sigma**2)

x = np.linspace(-4, 4, 100)
plt.figure(figsize = (4, 4))
plt.plot(x, gauss(0, 1, 1), 'k', linewidth = 3)
plt.plot(x, gauss(2, 3, 0.5), 'grey', linewidth = 3)
plt.ylim(-.5, 1.5)
plt.xlim(-4, 4)
plt.grid(1)
plt.show()
```
