- $f(x) = (x-2)x(x+2)$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(x):
    return (x-2)*x*(x+2)
x = np.random.randn(100)
x.sort()
y = f(x)
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f2(x, w):
    return (x-w)*x*(x+2)

# x를 정의
x = np.linspace(-3, 3, 100)

# 차트 묘사
plt.plot(x, f2(x, 2), color = 'black', label='$w=2$')
plt.plot(x, f2(x, 1), color = 'cornflowerblue', label='$w=1$')
plt.legend(loc="upper left") #(upper left, upper right, lower left, lower right)
plt.ylim(-15, 15) # y축 범위
plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```

```python
# 그래프에서 사용할 수 있는 색상 목록
import matplotlib
matplotlib.colors.cnames
```

```python
# 그래프를 여러개 보여주기
plt.figure(figsize=(10, 3)) # figure 지정
plt.subplots_adjust(wspace=0.5, hspace=0.5) # 그래프 간격 지정
colors = 'rgbcmy'
for i in range(6):
    plt.subplot(2, 3, i+1) # 그래프 위치
    plt.title(i+1)
    plt.plot(x, f2(x, i), list(colors)[i])
    plt.ylim(-20, 20)
    plt.grid(True)
plt.show()
```

- $f(x_0,x_1)=(2x_o^2+x_1^2)exp(-(2x_0^2+x_1^2))$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f3(x0, x1):
    r = 2*x0**2+x1**2
    answer = r*np.exp(-r)
    return answer

xn = 9
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
y = np.zeros((xn,xn))
for i in range(xn):
    for j in range(xn):
        y[j][i] = f3(x0[i], x1[j]) # j와 i 순서
print(x0)
print(np.round(y,1))
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f3(x0, x1):
    r = 2*x0**2+x1**2
    answer = r*np.exp(-r)
    return answer

xn = 9
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)
y = np.zeros((xn, xn))
for i in range(xn):
    y[:,i] = f3(x0[i],x1[:])
print(np.round(y,1))
```
