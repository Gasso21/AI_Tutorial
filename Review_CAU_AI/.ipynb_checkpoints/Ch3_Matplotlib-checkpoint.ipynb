- $f(x) = (x-2)x(x+2)$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(x):
    return (x-2)*x*(x+2)
x = np.random.randn(100)
x.sort()
y = f(x)
plt.plot(x, y)
plt.show()
```
