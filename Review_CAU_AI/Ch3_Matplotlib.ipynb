- $f(x) = (x-2)x(x+2)$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(x):
    return (x-2)*x*(x+2)
x = np.random.randn(100)
x.sort()
y = f(x)
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f2(x, w):
    return (x-w)*x*(x+2)

# x를 정의
x = np.linspace(-3, 3, 100)

# 차트 묘사
plt.plot(x, f2(x, 2), color = 'black', label='$w=2$')
plt.plot(x, f2(x, 1), color = 'cornflowerblue', label='$w=1$')
plt.legend(loc="upper left") #(upper left, upper right, lower left, lower right)
plt.ylim(-15, 15) # y축 범위
plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```

- **subplot**: 그래프를 여러개로

```python
# 그래프를 여러개 보여주기
plt.figure(figsize=(10, 3)) # figure 지정
plt.subplots_adjust(wspace=0.5, hspace=0.5) # 그래프 간격 지정
colors = 'rgbcmy'
for i in range(6):
    plt.subplot(2, 3, i+1) # 그래프 위치
    plt.title(i+1)
    plt.plot(x, f2(x, i), list(colors)[i])
    plt.ylim(-20, 20)
    plt.grid(True)
plt.show()
```

- $f(x_0,x_1)=(2x_o^2+x_1^2)exp(-(2x_0^2+x_1^2))$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f3(x0, x1):
    r = 2*x0**2+x1**2
    answer = r*np.exp(-r)
    return answer

xn = 9
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
y = np.zeros((xn,xn))
for i in range(xn):
    for j in range(xn):
        y[j][i] = f3(x0[i], x1[j]) # j와 i 순서
print(x0)
print(np.round(y,1))
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

def f3(x0, x1):
    r = 2*x0**2+x1**2
    ans = r*np.exp(-r)
    return ans

xn = 50
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)
xx0, xx1 = np.meshgrid(x0, x1)
y = np.zeros((xn, xn))

for i in range(xn):
    y[i, :] = f3(x0[:],x1[i])

fig = plt.figure(figsize = (5, 3.5))
ax = plt.subplot(1, 1, 1, projection = '3d')
ax.plot_surface(xx0, xx1, y, rstride=2, cstride=2, alpha=0.3, color='blue', edgecolor='black')
ax.set_zticks((0, 0.2))
ax.view_init(30, 45)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

- **contour**: 등고선

```python
xn = 50
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)

def f3(x0, x1):
    r = 2*x0**2+x1**2
    ans = r*np.exp(-r)
    return ans

y = np.zeros((xn, xn))
for i in range(xn):
    y[i,:] = f3(x0[:],x1[i])
    
xx0, xx1 = np.meshgrid(x0, x1)

plt.figure(1, figsize=(4, 4))
cont = plt.contour(xx0, xx1, y, 5, colors='black') # 등고선 높이를 5단계로 지정
cont.clabel(fmt='%3.2f', fontsize=8)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.show()
```
